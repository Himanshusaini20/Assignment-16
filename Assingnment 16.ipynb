### Q2 Create a simple Flask application to display 'Hello World!!'. Attach the screenshot of the output in the jupyter notebook.

```python
from flask import Flask
app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

### Q4 Create a"/welcome" route to display the welcome message "Wlcome to ABC Corporation " and a"/" route to show the following details:
Company Name; ABC Corporation
Location: India
Contact Detail: 999-99-9999
Attach the screeshot of the output in jupyter notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def details():
    return "<h1>Company Name: ABC Corporation</h1><h1>Location:India</h1><h1>Contact Detail: 99-999-9999</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
